```python
import pandas as pd
import numpy as np

retail_sales=pd.read_csv('retail_sales_data.csv')

print(retail_sales.info())
```

```python
retail_sales.head()
```

```python
retail_sales.nunique().to_frame(name='total_unique')
```

```python
retail_sales.isnull().sum()
```

```python
retail_sales['order_date']=pd.to_datetime(retail_sales['order_date'])
```

```python
retail_sales.info()
```

```python
retail_sales['discount']=retail_sales['discount'].fillna(0)
```

```python
retail_sales.isnull().sum()
```

```python
retail_sales['sales']=np.round(retail_sales['quantity'] * retail_sales['unit_price'] * (1-retail_sales['discount']/100),2)
```

```python

retail_sales['profit']=np.round(retail_sales['sales'] - retail_sales['quantity'] * retail_sales['cost'],2)
```

```python
retail_sales.head()
```

```python
retail_sales.to_csv('retail_sales_clean.csv',index=False)
print('Cleaned dataset saved as sales_data_cleaned.csv')
```
